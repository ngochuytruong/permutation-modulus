## Goal

We aim to solve the strictly convex quadratic program (primal)

$$ \min_{x\in \mathbb{R}^n}\;\frac{1}{2}\|x\|^2
  \qquad\text{subject to}\qquad
  Ax \ge \mathbb{1}, $$
where $A\in\mathbb{R}^{m\times n}$ and $\mathbb{1}\in\mathbb{R}^m$ is the all-ones vector.
The $i$-th constraint reads $a_i^{\top}x\ge 1$.



## How To Solve

The Lagrangian is $\mathcal{L}(x,\lambda) = \frac{1}{2}\|x\|^2 - \lambda^{\top}(Ax - \mathbb{1})$

With $\lambda \geq 0$

$\inf\limits_{x \in \mathbb{R}^n}(\sup\limits_{\lambda \in \mathbb{R}^m_{\geq 0}}(\mathcal{L}(x,\lambda))) = \sup\limits_{\lambda \in \mathbb{R}^m_{\geq 0}}(\inf\limits_{x \in \mathbb{R}^n}(\mathcal{L}(x,\lambda)))$ by Slater's condition

The left problem is the primal problem and the right problem is the dual problem

## Dual Problem

Now we look at the dual problem, lets find critical points with respect to $x$ in $$\mathcal{L}(x,\lambda) = \frac{1}{2}\|x\|^2 - \lambda^{\top}(Ax - \mathbb{1})$$

$\implies x = A^{\top}\lambda$, Plug in the critical point. Then we maximize over $\lambda$
$$ \max_{\lambda\ge 0}\; \phi(\lambda) := \mathbb{1}^{\top}\lambda - \frac{1}{2}\|{A^{\top}\lambda}\|^2.$$

$$\nabla\phi(\lambda) = \mathbb{1} - Ax.$$

KKT Conditions:
$$x = A^{\top}\lambda, \\
  Ax \ge \mathbb{1}, \\
  \lambda \ge 0, \\
  \lambda_i\bigl(a_i^{\top}x - 1\bigr) = 0,\quad i=1,\ldots,m.$$

  If $x \text{ and } \lambda$ satisfy KKT conditions then they are optimal

## Goldfarb-Idnani Algorithm

The algorithm will build a set of constraints by adding one at a time with the corresponding values of the Lagrangian dual variable $\lambda$.

We want these constraints to always be active. At times we will remove constraints that become inactive.

In particular, let W be the current active set. At all times we maintain that:

(i) $x = A_W^{\top}\lambda_W$ 

(ii) $A_Wx = \mathbb{1}$ 

(iii) $\lambda_W > 0$

A constraint $p$ is active if $a_p^{\top}x = 1$

## Searching

When starting the algorithm
$W = \emptyset, \lambda = \emptyset, x = \emptyset$

If the current choice of $\lambda$ is not optimal there must exist a constraint $p$ so that $a_p^{\top}x < 1.$ We add $p$ to $W$ and update $\lambda$ so that the 3 conditions for $W$ are fulfilled.

(i) $x = A_W^{\top}\lambda_W$ 

(ii) $A_Wx = \mathbb{1}$ 

(iii) $\lambda_W > 0$

We move $\lambda_W$ in the direction $\Delta\lambda_W$ with the amount $t$ $$\lambda_W(t) = \lambda_W + t\,\Delta\lambda_W$$

## We Choose A Direction $\Delta\lambda_W$ So That All Previous Constraints Remain Active

We parameterize the update as
$$
  \lambda_W(t) = \lambda_W + t\,\Delta\lambda_W,
  \qquad
  \lambda_p(t) = t,
  \qquad
  x(t) = x + t\,\Delta x,
$$

with $\Delta x = A_W^{\top}\Delta\lambda_W + a_p$.

Requiring $A_W\Delta x = 0$ to keep active constraints tight gives:


$$A_W A_W^{\top}\,\Delta\lambda_W = -A_W a_p$$

But solving this equation is an expensive step in our algorithm

## Choose $t$ So That The New Constraint Is Active

$$t_{\mathrm{full}}a_p^{\top}\Delta x + a_p^{\top}x = 1$$

When we are moving $\lambda$ we can make a previous $\lambda_j$ hit $0$ 

If that happens we would remove the constraint $j$ and recalculate the direction

If not we successfully add the constraint $p$

## Why Does It Work? - Dual Objective Does Not Decrease


**Lemma:** *No step decreases $\phi$.*

**Proof:**
$$

  \frac{d}{dt}\phi(\lambda(t))
  = (\mathbb{1} - Ax)^{\top}\Delta\lambda
  = \sum_{i\in W}\underbrace{(1 - a_i^{\top}x)}_{=\,0}\Delta\lambda_i
    + \underbrace{(1 - a_p^{\top}x)}_{>\,0}\underbrace{\Delta\lambda_p}_{=\,1}
  > 0.
$$
Active-constraint terms vanish because $a_i^{\top}x=1$ for $i\in W$.
Since $t > 0$, the objective strictly increases.
Dropping zero-multiplier indices does not change $\lambda$ or $x$,
so it does not change $\phi$. $\square$

## Termination

**Inner loop.**

Since $\lvert W\rvert$ is finite, the inner loop terminates.

**Outer loop.**

**Lemma** *No active set is ever revisited.*

Proof:

Linear independence of $A_W$ is maintained throughout: If a constraint $p$ is
added, then $$a_p^{\top}\Delta x = a_p^{\top}x_{\text{new}} - a_p^{\top}x_{\text{old}} = 1 - a_p^{\top}x_{\text{old}} > 0$$

If $a_p$ is a linear combination of rows $\in W$ then $a_p^{\top}\Delta x$ must be $0$ since all rows $\in W$ remain active before and after. So there is a contradiction.

So, $a_p \notin \operatorname{row}(A_W)$; dropping steps only remove rows.
Hence $A_W A_W^{\top}$ is invertible and $W$ uniquely determines
$$
  \lambda(W) = (A_W A_W^{\top})^{-1}\mathbb{1},
  \qquad x(W) = A_W^{\top}\lambda(W),
  \qquad \phi(\lambda(W)).
$$
Since $\phi$ strictly increases at every outer iteration
(by Lemma), the same value $\phi(\lambda(W))$---and hence
the same $W$---cannot recur. $\square$

Since there are at most $2^m$ distinct active sets, the algorithm terminates
in finitely many steps.

## Efficiency Step

Solving this equation: $$A_W A_W^{\top}\,\Delta\lambda_W = -A_W a_p$$ is an expensive step in our algorithm

If we know the Cholesky Decomposition of the square matrix $A_WA_W^{\top} = LL^{\top}$ where $L$ is a square lower triangular matrix then $\Delta\lambda_W$ can be found efficiently

Hence we update $L$ in every step efficiently (After deleting a row or adding a row to $A_W$)

## Constraint Addition

When all $\lambda_j > 0$ we add a constraint

This is done by adding the specific $a_p$ to our $A_W$

$$
\begin{bmatrix}
A_W \\ 
a_p^{\top} \\
\end{bmatrix}
\begin{bmatrix}
A_W^{\top} &
a_p
\end{bmatrix}
=
\begin{bmatrix}
A_W A_W^{\top} &
A_W a_p \\
a_p^{\top} A_W & 
a_p^{\top} a_p
\end{bmatrix}
$$
To update our Cholesky matrix $L$ as follows $\overline{L} = \begin{bmatrix}
L & 0 \\
l^{\top} & t
\end{bmatrix}$

So that

$$
\overline{L}{\overline{L}}^{\top} = 
\begin{bmatrix}
L & 0 \\
l^{\top} & t
\end{bmatrix}
\begin{bmatrix}
L^{\top} & l \\
0 & t
\end{bmatrix}
=
\begin{bmatrix}
LL^{\top} & Ll \\
l^{\top}L^{\top} & l^{\top}l + t^2
\end{bmatrix}
= 
\begin{bmatrix}
A_W A_W^{\top} &
A_W a_p \\
a_p^{\top} A_W & 
a_p^{\top} a_p
\end{bmatrix}
$$

Hence

$$l = L^{-1}A_Wa_p \text{ and } t = \sqrt{a_p^{\top}a_p - l^{\top}l}$$



## Constraint Removal

When a $\lambda_j$ <= 0 we remove all inactive constraints from the active set

This is done by removing the specific $a_p^{\top}$

$$
\begin{bmatrix}
A_{W_1} \\
a_p^{\top} \\
A_{W_2} \\
\end{bmatrix}
\begin{bmatrix}
A_{W_1}^{\top} & a_p & A_{W_2}^{\top}
\end{bmatrix}
=
\begin{bmatrix}
L_{11} & 0 & 0 \\
l_{21}^{\top} & t & 0 \\
L_{31} & l_{32} & L_{33}
\end{bmatrix}
\begin{bmatrix}
L_{11}^{\top} & l_{21} & L_{31}^{\top} \\
0 & t & l_{32}^{\top} \\
0 & 0 & L_{33}^{\top}
\end{bmatrix}
$$



Upon removing the offending constraint we have

$\overline{L} = \begin{bmatrix}
L_{11} & 0 & 0 \\
L_{31} & l_{32} & L_{33}
\end{bmatrix}$

$$
\begin{bmatrix}
A_{W_1} \\
A_{W_2} \\
\end{bmatrix}
\begin{bmatrix}
A_{W_1}^{\top}& A_{W_2}^{\top}
\end{bmatrix}
=
\begin{bmatrix}
L_{11} & 0 & 0 \\
L_{31} & l_{32} & L_{33}
\end{bmatrix}
\begin{bmatrix}
L_{11}^{\top} & L_{31}^{\top} \\
0 & l_{32}^{\top} \\
0  & L_{33}^{\top}
\end{bmatrix}
$$

This $\overline{L}$ has the dimension $i\times(i+1)$

And every nonzero entry above the new diagonal must be at the position $(j,j+1)$. 

To resolve this we perform a sequence of Givens rotations to update $\overline{L}$ to a square lower triangular form
