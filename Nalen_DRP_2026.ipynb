## Goal

Given this we aim to solve the strictly convex quadratic program

$$ \min_{x\in \R^n}\;\frac{1}{2}\|x\|^2
  \qquad\text{subject to}\qquad
  Ax \ge \mathbb{1}, $$
where $A\in\R^{m\times n}$ and $\mathbb{1}\in\R^m$ is the all-ones vector.
The $i$-th constraint reads $a_i^{\top}x\ge 1$.

## KKT Conditions

When solving the problem we fulfill KKT conditions to find optimal $\rho*$ and dual optimal $\lambda*$

These conditions are:
$$
x = A^{\top}\lambda,
$$
$$
Ax \ge \mathbb{1},
$$
$$
\lambda \ge 0,
$$
$$
\lambda_i\bigl(a_i^{\top}x - 1\bigr) = 0,\quad i=1,\ldots,m.
$$

## Initialization

When starting the algorithm with
$W$ be the set of active constraints

The $W = \empty, \lambda = \empty, x = \empty$

And so the previous KKT conditions are fulfilled vacuously
