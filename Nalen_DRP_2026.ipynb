## Goal

Given this we aim to solve the strictly convex quadratic program

$$ \min_{x\in \mathbb{R}^n}\;\frac{1}{2}\|x\|^2
  \qquad\text{subject to}\qquad
  Ax \ge \mathbb{1}, $$
where $A\in\mathbb{R}^{m\times n}$ and $\mathbb{1}\in\mathbb{R}^m$ is the all-ones vector.
The $i$-th constraint reads $a_i^{\top}x\ge 1$.

## How To Solve

The Langrangian is $\mathcal{L}(x,\lambda) = \frac{1}{2}\|x\|^2 - \lambda^{\top}(Ax - \mathbb{1})$

## Active Set

Let W be the current active set. At all times we maintain that:

(i) $x = A_W^{\top}\lambda_W$ 

(ii) $A_Wx = \mathbb{1}$ 

(iii) $\lambda_W > 0$

When constraints become inactive they have a $\lambda_W\leq0$. On each update we then remove all inactive constraints.

## Dual Problem

Since $x = A^{\top}\lambda$, the dual problem is
$$ \max_{\lambda\ge 0}\; \phi(\lambda) := \mathbb{1}^{\top}\lambda - \frac{1}{2}\|{A^{\top}\lambda}\|^2.$$

The primal variable is recovered by $x = A^{\top}\lambda $, and $$\nabla\phi(\lambda) = \mathbb{1} - Ax.$$

## Initialization

When starting the algorithm
$W = \emptyset, \lambda = \emptyset, x = \emptyset$

And so the invariant conditions are fulfilled vacuously

## Searching

If at any point a constriant $p$ is violated such that: $a_p^{\top} < 1.$ We update such that $\lambda_W(t) = \lambda_W + t\,\Delta\lambda_W$

## How Do we get $\Delta\lambda_W?$

$\Delta\lambda_W$ is evaluated by solving the equation $A_W A_W^{\top}\,\Delta\lambda_W = -A_W a_p$

But solving this equation is an expensive step in our algorithm

## Efficiency Step

To solve $A_W A_W^{\top}\,\Delta\lambda_W = -A_W a_p$ more efficiently we maintain a Cholesky Decomposition such that $A_W A_W^{\top} = LL^{\top}$

This allows us to solve for $\Delta\lambda_W$ much more efficiently

## Updating

When updating our active set we have two cases:

Case 1: Adding a constraint

This happens when all constraints are active and we have not terminated

Then a new constraint $p$ becomes active and $W \leftarrow W \cup {p}$

Case 2: Removal of all inactive constraints

This occurs if at least one $\lambda_j \leq 0$ We then remove all inactive constraints from $W$ so that $\lambda_W > 0$

In either case we continue to loop after the case is completed until the algorithm terminates

## Dual Objective Is Strictly Increasing


**Lemma:** *Every step strictly increases $\phi$.*

**Proof:**
$$

  \frac{d}{dt}\phi(\lambda(t))
  = (\mathbb{1} - Ax)^{\top}\Delta\lambda
  = \sum_{i\in W}\underbrace{(1 - a_i^{\top}x)}_{=\,0}\Delta\lambda_i
    + \underbrace{(1 - a_p^{\top}x)}_{>\,0}\underbrace{\Delta\lambda_p}_{=\,1}
  > 0.
$$
Active-constraint terms vanish because $a_i^{\top}x=1$ for $i\in W$.
Since $t > 0$, the objective strictly increases.
Dropping zero-multiplier indices does not change $\lambda$ or $x$,
so it does not change $\phi$. $\square$

## Termination

**Inner loop.**

Each blocking step has $t > 0$ (since $\lambda_W > 0$ by invariant (iii))
and drops at least one index from $W$.
Since $\lvert W\rvert$ is finite, the inner loop terminates.
Once $W = \emptyset$, $\Delta x = a_p$ and
$a_p^{\top}\Delta x = \|{a_p}\|^2 > 0$,
so $p$ is added.

**Outer loop.**

**Lemma** *No active set is ever revisited.*

Proof:

Linear independence of $A_W$ is maintained throughout: a constraint $p$ is
added only when $a_p^{\top}\Delta x > 0$, which implies
$a_p \notin \operatorname{row}(A_W)$; dropping steps only remove rows.
Hence $A_W A_W^{\top}$ is invertible and $W$ uniquely determines
$$
  \lambda(W) = (A_W A_W^{\top})^{-1}\mathbb{1},
  \qquad x(W) = A_W^{\top}\lambda(W),
  \qquad \phi(\lambda(W)).
$$
Since $\phi$ strictly increases at every outer iteration
(by Lemma), the same value $\phi(\lambda(W))$---and hence
the same $W$---cannot recur. $\square$

Since there are at most $2^m$ distinct active sets, the algorithm terminates
in finitely many steps.

## Constraint Addition

When all $\lambda_j > 0$ we add a constraint

This is done by adding the specific $a_p$ to our $A_W$

$$
\begin{bmatrix}
A_W \\ 
a_p^{\top} \\
\end{bmatrix}
\begin{bmatrix}
A_W^{\top} &
a_p
\end{bmatrix}
=
\begin{bmatrix}
A_W A_W^{\top} &
A_W a_p \\
a_p^{\top} A_W & 
a_p^{\top} a_p
\end{bmatrix}
$$
To update our Cholesky matrix $L$ we compute our new row $$l = L^{-1}A_Wa_p \text{ and } t = \sqrt{a_p^{\top}a_p - l^{\top}l}$$

Then
$$
\begin{bmatrix}
L & 0 \\
l^{\top} & t
\end{bmatrix}
\begin{bmatrix}
L^{\top} & l \\
0 & t
\end{bmatrix}
=
\begin{bmatrix}
LL^{\top} & Ll \\
l^{\top}L^{\top} & l^{\top}l + t^2
\end{bmatrix}
$$


## Constraint Removal

When a $\lambda_j$ <= 0 we remove all inactive constraints from the active set

This is done by removing the specific $a_p^{\top}$

$$
\begin{bmatrix}
A_{W_1} \\
a_p^{\top} \\
A_{W_2} \\
\end{bmatrix}
\begin{bmatrix}
A_{W_1}^{\top} & a_p & A_{W_2}^{\top}
\end{bmatrix}
=
\begin{bmatrix}
L_{11} & 0 & 0 \\
l_{21}^{\top} & t & 0 \\
L_{31} & l_{32} & L_{33}
\end{bmatrix}
\begin{bmatrix}
L_{11}^{\top} & l_{21} & L_{31}^{\top} \\
0 & t & l_{32}^{\top} \\
0 & 0 & L_{33}^{\top}
\end{bmatrix}
$$

Upon removing the offending constraint we have

$$
\begin{bmatrix}
A_{W_1} \\
A_{W_2} \\
\end{bmatrix}
\begin{bmatrix}
A_{W_1}^{\top}& A_{W_2}^{\top}
\end{bmatrix}
=
\begin{bmatrix}
L_{11} & 0 & 0 \\
L_{31} & l_{32} & L_{33}
\end{bmatrix}
\begin{bmatrix}
L_{11}^{\top} & L_{31}^{\top} \\
0 & l_{32}^{\top} \\
0  & L_{33}^{\top}
\end{bmatrix}
$$



## Givens Rotations

In the case where we remove a constraint our updated $L$ is no longer lower triangular

To resolve this we perform a series of Givens rotations to restore $L$ to its lower triangular status

Each Givens rotation only affects two columns of L each time allowing it to be done efficiently
