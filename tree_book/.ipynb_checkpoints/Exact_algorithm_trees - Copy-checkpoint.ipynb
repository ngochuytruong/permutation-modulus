```python
# Cell 1 — Imports & matplotlib setup
%matplotlib inline
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from itertools import combinations

# Optional: if you have modulus_tools installed in your environment
from modulus_tools.basic_algorithm import matrix_modulus, modulus
from modulus_tools.families.networkx_families import MinimumSpanningTree

plt.rcParams['figure.dpi'] = 130
plt.rcParams['font.size'] = 12
```

# Intro to Modulus on Graphs

This notebook introduces the notion of a **family of objects** on a graph, builds the corresponding **usage matrix** $N$, defines **admissible densities**, and computes **$p$-modulus** both (i) directly via a convex program and (ii) with the iterative basic algorithm from `modulus_tools`.


```python
# Cell 2 — Small demo graph (the "slashed house" style) and helpers
def slashed_house_graph():
    """
    A 5-node toy graph often used for small family examples.
    Nodes: 0,1,2,3,4 laid out like a house with one diagonal ("slash").
    """
    G = nx.Graph()
    edges = [
        (0,1), (1,2), (2,0),   # triangle base (0-1-2)
        (0,3), (1,3),          # roof connections
        (1,4), (2,4),          # right square side
        (3,4)                  # roof ridge (slash)
    ]
    G.add_edges_from(edges)
    pos = {
        0: (-1, 0), 1: (0, 0), 2: (1, 0),
        3: (-0.5, 1), 4: (0.7, 0.9)
    }
    return G, pos

def enumerate_edges(G):
    """Enumerate edges and store index at G[u][v]['enum']."""
    for i, (u, v) in enumerate(G.edges()):
        G[u][v]['enum'] = i

def usage_matrix_from_objects(G, list_of_edge_sets):
    """
    Build N from a list of objects; each object is an iterable of edges (u,v).
    Edges can be (u,v) or (v,u); we normalize with min/max for undirected graphs.
    """
    m = G.number_of_edges()
    N = np.zeros((len(list_of_edge_sets), m))
    for i, obj in enumerate(list_of_edge_sets):
        for (a, b) in obj:
            # Normalize edge orientation for undirected graphs
            if G.has_edge(a, b):
                j = G[a][b]['enum']
            else:
                j = G[b][a]['enum']
            N[i, j] = 1.0
    return N

G, pos = slashed_house_graph()
enumerate_edges(G)
m = G.number_of_edges()
m, list(G.edges(data=True))
```

## 1) Families of objects and the usage matrix $N$

A **family** $\Gamma$ is a collection of “objects” (paths, cuts, cycles, triangles, spanning trees, …).  
For many families, each object can be identified with the subset of edges it uses.  

The **usage matrix** $N \in \mathbb{R}_{\ge 0}^{|\Gamma| \times m}$ (with $m=|E|$) records, row-by-row, which edges each object uses.


### Example A — Triangles family

We build $\Gamma =$ set of triangles and the corresponding usage matrix $N_{\triangle}$.  
We also visualize each triangle with edge enumeration.


```python
# Cell 3 — Triangles: build family and N
def all_triangles(G):
    """Return list of triangles as lists of edges."""
    triangles = []
    nodes = list(G.nodes())
    n = len(nodes)
    index = {v:i for i,v in enumerate(nodes)}
    # brute-force triples
    for i in range(n-2):
        for j in range(i+1, n-1):
            for k in range(j+1, n):
                a, b, c = nodes[i], nodes[j], nodes[k]
                if G.has_edge(a,b) and G.has_edge(b,c) and G.has_edge(c,a):
                    triangles.append([(a,b),(b,c),(c,a)])
    return triangles

triangles = all_triangles(G)
N_tri = usage_matrix_from_objects(G, triangles)

# Draw each triangle and print N
plt.figure(figsize=(10,4))
labels = {(u, v): d['enum'] for u, v, d in G.edges(data=True)}
for i, T in enumerate(triangles):
    plt.subplot(1, max(1,len(triangles)), i+1)
    nx.draw(G, pos, node_size=100, node_color='black', edge_color='lightgray')
    nx.draw_networkx_edges(G, pos, edgelist=T, width=3)
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_size=11)
    plt.title(f"Triangle {i+1}")
plt.tight_layout()

print("N_tri =\n", N_tri)
```

### Example B — Simple paths between two nodes

Pick two nodes $s,t$ and collect all **simple paths** between them.  
Construct $N_{\text{path}}$ and visualize.


```python
# Cell 4 — Paths: build family and N
s, t = 1, 3  # top-ish to left-roof node in this toy layout
paths = list(nx.all_simple_paths(G, s, t))

# Edge lists for each path
path_edge_sets = []
for path in paths:
    edges = [(path[i], path[i+1]) for i in range(len(path)-1)]
    path_edge_sets.append(edges)

N_path = usage_matrix_from_objects(G, path_edge_sets)

# Draw each path and print N
rows = int(np.ceil(len(paths)/3))
plt.figure(figsize=(10, 3.2*rows))
for i, path in enumerate(paths):
    plt.subplot(rows, 3, i+1)
    edges = [(path[k], path[k+1]) for k in range(len(path)-1)]
    nx.draw(G, pos, node_size=100, node_color='black', edge_color='lightgray')
    nx.draw_networkx_edges(G, pos, edgelist=edges, width=3)
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_size=11)
    plt.title(f"Path {i+1}: {path}")
plt.tight_layout()

print("N_path =\n", N_path)
```

## 2) Admissible densities

A **density** $\rho \in \mathbb{R}_{\ge 0}^{m}$ is **admissible** for $\Gamma$ if every object pays at least $1$:
\[
N \rho \;\;\ge\;\; \mathbf{1}.
\]

Below we draw random $\rho$ and color each object green if its $\rho$-length $(N\rho)_i \ge 1$, else red.


```python
# Cell 5 — Admissibility demos (triangles and paths)
rng = np.random.default_rng(88198283)

def plot_admissibility_on_triangles(G, triangles, N_tri, n_rho=3):
    n_edges = G.number_of_edges()
    for i in range(n_rho):
        rho = rng.random(n_edges)
        ell = N_tri @ rho

        plt.figure(figsize=(10,4))
        labs = {(u, v): round(rho[d['enum']], 3) for u, v, d in G.edges(data=True)}
        for j, T in enumerate(triangles):
            plt.subplot(1, max(1,len(triangles)), j+1)
            good = (ell[j] >= 1 - 1e-12)
            color = 'green' if good else 'red'
            nx.draw(G, pos, node_size=100, node_color='black', edge_color='lightgray')
            nx.draw_networkx_edges(G, pos, edgelist=T, width=3, edge_color=color)
            nx.draw_networkx_edge_labels(G, pos, edge_labels=labs, font_size=11)
            plt.title(f"ℓ_j = {ell[j]:.3f}")
        title = f"Example {i+1} — {'Admissible' if np.all(ell>=1) else 'Not admissible'}"
        plt.suptitle(title)
        plt.tight_layout()
        plt.subplots_adjust(top=0.82)

def plot_admissibility_on_paths(G, paths, N_path, n_rho=3):
    n_edges = G.number_of_edges()
    for i in range(n_rho):
        rho = rng.random(n_edges)
        ell = N_path @ rho

        rows = int(np.ceil(len(paths)/3))
        plt.figure(figsize=(10, 3.2*rows))
        labs = {(u, v): round(rho[d['enum']], 3) for u, v, d in G.edges(data=True)}
        for j, path in enumerate(paths):
            plt.subplot(rows, 3, j+1)
            edges = [(path[k], path[k+1]) for k in range(len(path)-1)]
            good = (ell[j] >= 1 - 1e-12)
            color = 'green' if good else 'red'
            nx.draw(G, pos, node_size=100, node_color='black', edge_color='lightgray')
            nx.draw_networkx_edges(G, pos, edgelist=edges, width=3, edge_color=color)
            nx.draw_networkx_edge_labels(G, pos, edge_labels=labs, font_size=11)
            plt.title(f"ℓ_j = {ell[j]:.3f}")
        title = f"Example {i+1} — {'Admissible' if np.all(ell>=1) else 'Not admissible'}"
        plt.suptitle(title)
        plt.tight_layout()
        plt.subplots_adjust(top=0.9)

plot_admissibility_on_triangles(G, triangles, N_tri, n_rho=2)
plot_admissibility_on_paths(G, paths, N_path, n_rho=2)
```

## 3) $p$-Modulus via convex optimization

Given $N$ and $p \in [1,\infty]$, the $p$-**energy** of $\rho$ is
\[
\mathcal{E}_p(\rho) =
\begin{cases}
\sum_e \rho_e^p, & p < \infty, \\
\|\rho\|_\infty, & p = \infty.
\end{cases}
\]

The **$p$-modulus** of $\Gamma$ is
\[
\operatorname{Mod}_p(\Gamma) \;=\; \min_{\rho \ge 0} \;\mathcal{E}_p(\rho)
\quad \text{s.t. } N \rho \ge \mathbf{1}.
\]

We can compute this directly with `matrix_modulus(N, p)`.


```python
# Cell 6 — Triangle modulus for several p values (using matrix_modulus)
p_list = (1, 1.5, 2, 2.5, 3, np.inf)

plt.figure(figsize=(10,8))
for i, p in enumerate(p_list):
    mod_val, rho_opt, _ = matrix_modulus(N_tri, p)

    plt.subplot(2,3,i+1)
    labs = {(u, v): round(rho_opt[d['enum']], 3) for u, v, d in G.edges(data=True)}
    nx.draw(G, pos, node_size=100, node_color='black', edge_color='lightgray')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labs, font_size=11)
    plt.title(f"Triangles: p = {p},  Mod ≈ {mod_val:.3f}")
plt.tight_layout()
```

## 4) Computing spanning-tree modulus

A key family is $\Gamma =$ all **spanning trees** of $G$.  
We will:

1. Build the usage matrix $N_{\text{tree}}$ by enumerating spanning trees.  
2. Compute modulus via `matrix_modulus`.  
3. Use the **basic algorithm** (`modulus`) with `MinimumSpanningTree` as the shortest-object oracle.

> Note: Enumerating spanning trees by brute force scales poorly. Keep graphs small for the matrix method.


## 5) The basic algorithm (scales to larger graphs)

The iterative method `modulus(m, matrix_modulus, shortest_oracle, …)` avoids building $N$.  

We give it:

- $m = |E|$,  
- a subproblem solver (`matrix_modulus`) for a current subfamily,  
- a shortest-object oracle (here `MinimumSpanningTree`).  

It returns $(\text{modulus\_value}, \text{constraints}, \rho, \lambda)$; we visualize $\rho$ as edge weights.


```python
# Cell 7 — Build the spanning-tree usage matrix N_tree (small graphs)
def spanning_tree_usage_matrix(G):
    """
    Builds N for the family of spanning trees by brute force over edge subsets.
    WARNING: exponential in general; use only for small graphs.
    """
    enumerate_edges(G)  # ensure 'enum'
    n = G.number_of_nodes()
    edges = list(G.edges())
    m = len(edges)
    rows = []
    for T_edges in combinations(edges, n-1):
        H = nx.Graph()
        H.add_nodes_from(G.nodes())  # keep same node set
        H.add_edges_from(T_edges)
        if nx.is_tree(H):
            v = np.zeros(m)
            for (a,b) in T_edges:
                j = G[a][b]['enum']
                v[j] = 1.0
            rows.append(v)
    if not rows:
        return np.zeros((0, m))
    return np.vstack(rows)

# Tiny demo graph: a 6-cycle with one diagonal
H = nx.cycle_graph(6)
H.add_edge(0, 3)
posH = nx.spring_layout(H, seed=42)
enumerate_edges(H)
N_tree = spanning_tree_usage_matrix(H)
print(f"H has {H.number_of_nodes()} nodes, {H.number_of_edges()} edges")
print(f"Number of spanning trees included in N_tree: {N_tree.shape[0]}")
print("Shape of N_tree:", N_tree.shape)
```

## 6) (Optional) Direct CVXPy formulation for small $N$

For completeness, here is a minimal CVXPy solver for
\[
\min_{\rho \ge 0} \; \sum_e \rho_e^p 
\quad \text{s.t. } N \rho \ge \mathbf{1},
\]
when $p \ge 1$ is finite.  

Use this only for small problems (you already have `matrix_modulus`).


```python
# Cell 8 — Spanning-tree modulus via matrix_modulus (small graph only)
for p in (2, 3, np.inf):
    mod_val, rho_opt, _ = matrix_modulus(N_tree, p)
    plt.figure(figsize=(4.2,3.6))
    labs = {(u, v): round(rho_opt[d['enum']], 3) for u, v, d in H.edges(data=True)}
    nx.draw(H, posH, node_size=80, node_color='black', edge_color='lightgray', width=1.5)
    nx.draw_networkx_edge_labels(H, posH, edge_labels=labs, font_size=10)
    plt.title(f"Spanning Trees (p={p})\nMod ≈ {mod_val:.4f}")
    plt.tight_layout()
```

```python
# Cell 9 — Helper that runs the basic algorithm and plots rho (edge colors)
def run_basic_algorithm_and_plot(G, layout_seed=123, max_iter=400, title=""):
    pos = nx.spring_layout(G, seed=layout_seed)
    m = G.number_of_edges()
    mst = MinimumSpanningTree(G)
    mod_val, cons, rho, lam = modulus(m, matrix_modulus, mst, max_iter=max_iter)

    plt.figure(figsize=(4.5,4.2))
    nx.draw(G, pos, node_size=30, node_color='black',
            width=2, edge_color=rho, edge_cmap=plt.cm.Set2)
    plt.title(f"{title}\nMod ≈ {mod_val:.4f}")
    plt.tight_layout()
    return rho, mod_val

# Demo 1 — random geometric graph (small)
RG = nx.random_geometric_graph(12, 0.55, seed=3810312)
rho_rg, mod_rg = run_basic_algorithm_and_plot(RG, title="Random geometric graph — Spanning-tree modulus")

# Demo 2 — a “ladder + ring” style graph
P = nx.complete_graph(6)
for u in range(6):
    P.add_edge(u, u+6)
nx.add_path(P, [6,7,8,9,10,11,6])
rho_p, mod_p = run_basic_algorithm_and_plot(P, title="Hybrid graph — Spanning-tree modulus")

print("Example rho ratios (η* = rho / Mod):")
print("RG:", np.round(rho_rg / mod_rg, 3))
print("Hybrid:", np.round(rho_p / mod_p, 3))
```

```python
# Cell 10 — Minimal CVXPy solve for finite p (uses SCS)
import cvxpy as cvx

def modulus_via_cvxpy(N, p=2.0):
    """
    Solve min sum(rho^p) s.t. N rho >= 1, rho >= 0.
    Finite p >= 1 only. For p = np.inf, prefer matrix_modulus which handles it internally.
    """
    n_edges = N.shape[1]
    rho = cvx.Variable(n_edges, nonneg=True)
    constraints = [N @ rho >= 1]
    if np.isfinite(p):
        obj = cvx.Minimize(cvx.pnorm(rho, p)**p)
    else:
        raise ValueError("Use finite p or call matrix_modulus for p = ∞.")
    prob = cvx.Problem(obj, constraints)
    prob.solve(solver=cvx.SCS, verbose=False)
    return prob.value, rho.value

# Quick check: triangle modulus at p=2 via CVXPy
val_cvx, rho_cvx = modulus_via_cvxpy(N_tri, p=2.0)
print("CVXPy (triangles, p=2): Mod ≈", float(val_cvx))
print("rho* ≈", np.round(rho_cvx, 3))
 
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import networkx as nx
from modulus_tools.basic_algorithm import matrix_modulus, modulus
from modulus_tools.families.networkx_families import MinimumSpanningTree
import numpy as np
import cvxpy as cvx
from itertools import combinations
import random
```

Define the usage matrix of spanning tree families

```python
def matrix(G):
    for i,(u,v) in enumerate(G.edges()):
        G[u][v]['enum'] = i
    n = len(G.nodes)
    N=[]
    for T in combinations(G.edges,n-1):
            H = nx.Graph(T)
            if nx.is_tree(H):   
                v = np.zeros(m)
                for (a,b) in T:
            
                    ind = G[a][b]['enum']
                    v[ind] = 1
                N.append(v)       
    return(np.array(N))
```

Compute modulus using the definition

```python
def modul(N,p,sigma):
    # number of rho variables
    n = N.shape[-1]
    # make sigma into a diagonal matrix
    S = np.diag(sigma)
    # set up optimization variables
    rho = cvx.Variable(n)
    X= (S**(1./p)) @ rho
    obj = cvx.Minimize(cvx.pnorm(X,p)**p)
    cons = [rho>=0,N @ rho >=1 ]
    # solve the problem
    prob = cvx.Problem(obj,cons)
    prob.solve(solver=cvx.SCS)
    return  np.array(obj.value).flatten(), np.array(rho.value).flatten()
```

Example 1

```python
G = nx.random_geometric_graph(8, 0.7, seed=3810312)
pos = {v:d['pos'] for v,d in G.nodes(data=True)}
m = len(G.edges())
matrix(G)
mod,rho= modul(matrix(G),2,np.ones(m))
#compute eta^* for spanning tree modulus
eta = np.round(rho/mod,3)
print(eta)
plt.figure(figsize=(5,5))
nx.draw(G, pos, node_size=50, node_color='black', width=2, edge_color=eta, edge_cmap=plt.cm.Set2)
```

Compute modulus using the basic algorithm

```python
def MODD(G):
    pos = nx.spring_layout(G,seed = 123)
    m = len(G.edges())
    mst = MinimumSpanningTree(G)
    mod, cons, rho, lam = modulus(m, matrix_modulus, mst, max_iter=400)
    plt.figure(figsize=(5,5))
    nx.draw(G, pos, node_size=10, node_color='black', width=2, edge_color=rho, edge_cmap=plt.cm.Set2)
    return rho,mod
```

```python
P = nx.complete_graph(6)
for u in range(6):
    P.add_edge(u,u+6)
nx.add_path(P,[6,7,8,9,10,11,6])
rho,mod=MODD(P)
#compute eta^* for spanning tree modulus
eta=np.round(rho/mod,3)
print(eta)
```
